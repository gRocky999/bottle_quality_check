```python
from google.colab import files
import cv2
from ultralytics import YOLO
import os

# Step 1: Upload your input video
uploaded = files.upload()  # e.g. close1.avi
input_video = next(iter(uploaded))

# Step 2: Load trained YOLOv8 model
model = YOLO("runs/detect/yolov8n_bottle_defects/weights/best.pt")

# Step 3: Run inference with smart label placement
results = model.predict(
    source=input_video,
    save=False,
    stream=True,
    conf=0.4
)

# Step 4: Set up video writer
cap = cv2.VideoCapture(input_video)
fps = cap.get(cv2.CAP_PROP_FPS)
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter("close1_output.mp4", fourcc, fps, (w, h))

# Color palette
def get_color(cls):
    palette = [(255, 99, 71), (60, 179, 113), (100, 149, 237), (218, 112, 214), (255, 215, 0)]
    return palette[cls % len(palette)]

# Prevent overlapping labels
def avoid_label_overlap(label_positions, y_start, spacing=25):
    while y_start in label_positions:
        y_start += spacing
    label_positions.add(y_start)
    return y_start

# Process each frame and beautify annotations
for result in results:
    frame = result.orig_img.copy()
    label_positions = set()

    for box in result.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        cls_id = int(box.cls[0])
        label = model.names[cls_id]
        conf = float(box.conf[0])
        color = get_color(cls_id)

        label_text = f"{label} {conf:.2f}"
        (tw, th), _ = cv2.getTextSize(label_text, cv2.FONT_HERSHEY_SIMPLEX, 0.6, 1)

        # Smart label Y-position
        ideal_y = y1 - 10 if y1 - 10 > 20 else y1 + 25
        label_y = avoid_label_overlap(label_positions, ideal_y)

        # Draw box
        cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
        # Draw filled background and label
        cv2.rectangle(frame, (x1, label_y - th - 6), (x1 + tw + 8, label_y + 4), color, -1)
        cv2.putText(frame, label_text, (x1 + 4, label_y), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1)

    out.write(frame)

cap.release()
out.release()

# Step 5: Download output video

files.download("close1_output.mp4")
```
